### Salary prediction from Random numbers using linear regression


Taking 500 random values between 1 and 50 and making prediction of salary based on the number of skills a person has

```python
import numpy as np
import pandas as pd
```

```python
skills = np.linspace(1, 50, 500, dtype='int')
```

```python
print(skills)
```

```python
salaries = skills* 8000
```

```python
buffer = np.random.randint(1,40000,500)
```

```python
buffer
```

```python
salaries = (skills* 8000)+buffer
```

```python
salaries
```

```python
df = pd.DataFrame({
    "Skills": skills, 
    "Salaries":salaries})
```

```python
df
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df[["Skills"]].values
Y=df[["Salaries"]].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size= 0.8)
```

```python
X_train.shape
```

```python
Y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linReg = LinearRegression()
```

```python
linReg.fit(X_train, Y_train)
```

```python
linReg.intercept_
```

```python
usk=int(input("Enter number of skills and i will help you predict your salary"))
print("Your predicted salary is :" ,linReg.predict([[usk]])[0][0])
```
